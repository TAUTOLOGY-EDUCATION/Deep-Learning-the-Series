```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_addons as tfa
from tensorflow.keras.layers import Dense, PReLU
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD
```

# ========== Data ==========

```python
data = pd.read_csv('dataset.csv')
```

```python
data
```

```python
X = data[['x1', 'x2']]
```

```python
y = data['y']
```

# ========== Model ==========

### === Setting Parameters ===

```python
input_dim = X.shape[1]
epochs = 500
learning_rate = 0.01
number_of_hidden_layers = 2
number_of_nodes_in_hidden_layers = [5, 5]
activation_function = ['relu', 'relu']
```

### === Create Model ===

```python
model = Sequential()

model.add(
    Dense(
        number_of_nodes_in_hidden_layers[0],
        input_dim=input_dim,
        activation=activation_function[0]
    )
)

for i, number_of_node in enumerate(number_of_nodes_in_hidden_layers[1:], start=1):
    model.add(
        Dense(
            number_of_node,
            activation=activation_function[i]
        )
    )

model.add(
    Dense(
        1, 
        activation=None
    )
)
```

### === Compile Model ===

```python
optimizer = SGD(
    learning_rate=learning_rate
)

model.compile(
    loss=MeanSquaredError(),
    optimizer=optimizer
)
```

```python
tf.config.list_physical_devices()
```

```python
device_name = '/device:CPU:0'
```

### === Train Model ===

```python
tqdm_callback = tfa.callbacks.TQDMProgressBar(show_epoch_progress=False)

with tf.device(device_name):
    history = model.fit(
        X,
        y,
        epochs=epochs,
        callbacks=[tqdm_callback],
    )
```

```python
plt.plot(history.history['loss'], label='training loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
```
