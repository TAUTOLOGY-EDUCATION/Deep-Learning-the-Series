```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_addons as tfa
from tensorflow.keras.layers import Dense, PReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import SGD, Adam

from sklearn.preprocessing import OneHotEncoder

np.random.seed(12345)
tf.random.set_seed(12345)
```

# ========== Data ==========

```python
data = pd.read_csv('dataset3.csv')
```

```python
data
```

```python
X = data[['x1', 'x2']]
```

```python
y = data['y']
```

```python
enc = OneHotEncoder(sparse=False, handle_unknown='ignore')
y_enc = enc.fit_transform(y.to_numpy().reshape(-1, 1))
```

# ========== Model ==========

```python
tf.config.list_physical_devices()
```

```python
devices_type = [device.device_type for device in tf.config.list_physical_devices()]

devices_type
```

```python
if 'GPU' in devices_type:
    device_name = '/device:GPU:0'
else:
    device_name = '/device:CPU:0'
```

```python
model = Sequential()

model.add(Dense(5, input_dim=X.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(y_enc.shape[1], activation='softmax'))

model.compile(loss=CategoricalCrossentropy(), optimizer=Adam(learning_rate=1))
```

```python
tqdm_callback = tfa.callbacks.TQDMProgressBar(show_epoch_progress=False)

with tf.device(device_name):
    history = model.fit(X, y_enc, epochs=500, batch_size=32, callbacks=[tqdm_callback])
```

```python
plt.plot(history.history['loss'], label='traning_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
```

```python
model.get_weights()
```
