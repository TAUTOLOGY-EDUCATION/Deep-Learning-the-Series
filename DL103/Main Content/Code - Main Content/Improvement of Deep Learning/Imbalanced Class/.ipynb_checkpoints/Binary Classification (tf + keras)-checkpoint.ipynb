```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_addons as tfa
from tensorflow.keras.layers import Dense, PReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import SGD

from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight

import warnings
warnings.filterwarnings('ignore')

np.random.seed(12345)
tf.random.set_seed(12345)
```

# ========== Data ==========

```python
data = pd.read_csv('dataset.csv')
```

```python
data
```

```python
X = data[['x1', 'x2']]
```

```python
y = data['y']
```

```python
le = LabelEncoder()
y_le = le.fit_transform(y)
```

# ========== Model ==========

```python
tf.config.list_physical_devices()
```

```python
devices_type = [device.device_type for device in tf.config.list_physical_devices()]

devices_type
```

```python
if 'GPU' in devices_type:
    device_name = '/device:GPU:0'
else:
    device_name = '/device:CPU:0'
```

```python
model = Sequential()

model.add(Dense(5, input_dim=X.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss=BinaryCrossentropy(), optimizer=SGD(learning_rate=1))
```

```python
class_weight = compute_class_weight('balanced', np.unique(y), y)
class_weight = dict(enumerate(class_weight.flatten()))

tqdm_callback = tfa.callbacks.TQDMProgressBar(show_epoch_progress=False)

with tf.device(device_name):
    history = model.fit(X, y_le, epochs=500, class_weight=class_weight, callbacks=[tqdm_callback])
    
class_weight
```

```python
plt.plot(history.history['loss'], label='traning_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
```

```python
model.get_weights()
```

# ========== Prediction ==========

```python
X = pd.read_csv('incoming_data.csv')
```

```python
model.predict(X)
```
