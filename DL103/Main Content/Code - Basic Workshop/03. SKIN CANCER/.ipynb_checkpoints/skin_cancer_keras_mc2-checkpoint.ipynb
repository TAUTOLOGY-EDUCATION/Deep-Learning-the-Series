# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import (
    OrdinalEncoder, 
    OneHotEncoder, 
    StandardScaler, 
    MinMaxScaler
)
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report

import warnings
warnings.filterwarnings('ignore')

np.random.seed(12345)
```

```python
import os
import sys

lib_path = os.path.abspath('../../')
sys.path.append(lib_path)

from easy_keras import KerasMLPClassifier, make_confusion_matrix
```

```python
seed = 12345

random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
```

# 2. Read Data

```python
data = pd.read_excel('job_review.xlsx')
```

```python
# ใช้สำรับใส่ภาพประกอบ
data
```

```python
data[:10]
```

# 3. Clean Data

## 3.1. Handle Missing Values

### 3.1.1. Check Missing Values

```python
data.info()
```

### 3.1.2. Remove Missing Values

```python
data.dropna(axis=0, inplace=True)
```

## 3.2. Handle Outliers

### 3.2.1. Check Outliers

```python
data.describe()
```

### 3.2.2. Remove Outliers

```python
_filter = data['salary'] > 0
data = data[_filter]
```

# 4. Split Dataset into Training & Test Sets

```python
target_name = 'score'
feature_name = list(data.columns.drop(target_name))
```

```python
X = data[feature_name]
y = data[target_name]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True)
```

# 5. Data Preparation

```python
numerical_feature = ['experience', 'salary']
categorical_feature = ['position', 'working_hour']
```

```python
for feature in categorical_feature:
    print(feature, ':', np.unique(X_train[feature]))
```

```python
ordinal_feature = ['working_hour']
```

```python
nominal_feature = ['position']
```

## 5.1. Ordinal Encoding

```python
### ใช้สำรับใส่ภาพประกอบ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True)
X_train[ordinal_feature].reset_index()
```

```python
categories = [
    np.array(['3-4', '5-7', '8-9', '10-12'], dtype=object)
]
```

### 5.1.1. Training Set

```python
ordinal_encoder = OrdinalEncoder(categories=categories)
X_train[ordinal_feature] = ordinal_encoder.fit_transform(X_train[ordinal_feature])
```

### 5.1.2. Test Set

```python
X_test[ordinal_feature] = ordinal_encoder.transform(X_test[ordinal_feature])
```

```python
### ใช้สำรับใส่ภาพประกอบ
X_train[ordinal_feature].reset_index()
```

## 5.2. One Hot Encoding

```python
### ใช้สำรับใส่ภาพประกอบ
X_train[nominal_feature].reset_index()
```

### 5.2.1. Training Set

```python
one_hot_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
one_hot_encoder.fit(X_train[nominal_feature])
```

```python
one_hot_feature = []
for i, feature in enumerate(nominal_feature):
    for cate in one_hot_encoder.categories_[i]:
        one_hot_feature_name = str(feature) + '_' + str(cate)
        one_hot_feature.append(one_hot_feature_name)
```

```python
X_train[one_hot_feature] = one_hot_encoder.transform(X_train[nominal_feature])
X_train.drop(nominal_feature, axis=1, inplace=True)
```

### 5.2.2. Test Set

```python
X_test[one_hot_feature] = one_hot_encoder.transform(X_test[nominal_feature])
X_test.drop(nominal_feature, axis=1, inplace=True)
```

```python
### ใช้สำรับใส่ภาพประกอบ
X_train[one_hot_feature].reset_index()
```

## 5.3. Feature Scaling

### 5.3.1. Training Set

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

### 5.3.2. Test Set

```python
X_test_scaled = scaler.transform(X_test)
```

```python
### ใช้สำรับใส่ภาพประกอบ
pd.DataFrame(X_train_scaled, columns=['experience', 'salary', 'working_hour', 'position_accountant', 'position_engineer', 'position_secretary'])
```

# 6. Model Creation

## 6.1. Setting Parameters

```python
classes = np.unique(y_train)
```

```python
clf = KerasMLPClassifier(
    input_dim=X_train_scaled.shape[1],
    hidden_layer_sizes=(25, 5),
    activation_function='relu',
    classes=classes,
    learning_rate_init=1,
    epochs=1000
)
```

## 6.2. Train Model

```python
clf.fit(X_train_scaled, y_train)
```

## 6.3. Loss Curve

```python
plt.plot(clf.loss_curve_, label='training_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
```

## 6.4. Model's Weight & Bias

```python
for coef_ in clf.coefs_:
    print(coef_.shape)
```

```python
for intercept_ in clf.intercepts_:
    print(intercept_.shape)
```

# 7. Prediction

## 7.1. Training Set

```python
y_pred_train = clf.predict(X_train_scaled)
```

## 7.2. Test Set

```python
y_pred_test = clf.predict(X_test_scaled)
```

# 8. Model Evaluation

## 8.1. Training Set

### 8.1.1. Confusion Matrix

```python
cm = confusion_matrix(y_train, y_pred_train)
make_confusion_matrix(
    cm,
    cmap='viridis',
    categories=classes,
    sum_stats=False,
    percent=False,
)
plt.show()
```

### 8.1.2. Scoring

```python
report = classification_report(y_train, y_pred_train, output_dict=True)
```

```python
print('accuracy =', report['accuracy'])
```

```python
pd.DataFrame.from_dict(report).T
```

## 8.2. Test Set

### 8.2.1. Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred_test)
make_confusion_matrix(
    cm,
    cmap='viridis',
    categories=classes,
    sum_stats=False,
    percent=False,
)
plt.show()
```

### 8.2.2. Scoring

```python
report = classification_report(y_test, y_pred_test, output_dict=True)
```

```python
print('accuracy =', report['accuracy'])
```

```python
pd.DataFrame.from_dict(report).T
```
