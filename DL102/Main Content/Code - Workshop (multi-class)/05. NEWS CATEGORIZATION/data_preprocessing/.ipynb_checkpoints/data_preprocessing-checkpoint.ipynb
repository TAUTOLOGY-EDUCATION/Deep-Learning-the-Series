```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

np.random.seed(123456)
```

```python
data = pd.read_csv('prachathai-67k.csv')

data.drop(['url', 'date'], axis=1, inplace=True)
data['labels'] = data['labels'].apply(eval)

data
```

```python
all_labels = pd.Series([x for _list in data['labels'] for x in _list])
value_counts = all_labels.value_counts()

all_tag_count = value_counts.values.sum()
tag_retios = np.array([])
for i in range(100):
    tag_retio = value_counts[:i].values.sum() / all_tag_count
    tag_retios = np.append(tag_retios, tag_retio)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(
    tag_retios,
)
ax.set_ylabel('Frequency retio')
ax.set_xlabel('First unique tag size')
ax.set_xticks(np.arange(0, 110, 10))
plt.show()
```

```python
value_counts[:15].keys()
```

```python
# select 15 first tag
selected_keys = value_counts[:15].keys()
bool_dict = {}
for i, item in enumerate(selected_keys):
    bool_dict[item] = data['labels'].apply(lambda x: item in x)
labels = pd.DataFrame(bool_dict)

labels
```

```python
labels = labels.drop(columns=['ข่าว', 'บทความ', 'รายงานพิเศษ'])

has_tag_index = labels.sum(axis=1) == 1
labels = labels[has_tag_index]
data = data[has_tag_index]

labels
```

```python
labels.columns = [
    'politics',
    'human_rights',
    'quality_of_life',
    'international',
    'social',
    'environment',
    'economy',
    'culture',
    'labor',
    'stability',
    'ict',
    'education',
]

true_indices = np.argwhere(labels.to_numpy())[:, 1]
data['labels'] = labels.columns[true_indices]

data.to_excel('news_dataset.xlsx', index=False)
data.to_excel('../news_dataset.xlsx', index=False)

data
```
